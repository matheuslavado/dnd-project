### Alterando coluna de tendências

```python
# Arrumando coluna de tendencias
def new_tendencia(x):
    try:
        ten = re.findall(r'([a-záéíóúçãõêôâ\s]+) ?[A-Z]', x)[0]
        return ten
    except:
        return np.nan
```

```python
data['Tendência'] = data['Tendência'].apply(lambda x: new_tendencia(x))
```

### Corrigindo coluna 'PV_dados'

```python
# corrigindo pv em relação aos sinais de '+' e '-'

def new_pv(x):
    if '+' not in x:
        x = x.replace(' ','-')
        if '--' in x:
            x = x.replace('--', ' - ')
        else:
            x = x
    else:
        x = x
        
    return x
```

```python
data['PV_dados'] = data['PV_dados'].apply(lambda x: new_pv(x))
```

```python
data['PV_dados'].unique()
```

## Coluna 'Deslocamento'

```python
data['Deslocamento'].unique()
```

### Corrigindo coluna 'Deslocamento'

```python
# tratando coluna de deslocamento

def new_desc(x):
    x = x.replace(',','.')
    x = re.findall('[\d.]+', x)[0]
    x = float(x)
    return x
```

```python
data['Deslocamento'] = data['Deslocamento'].apply(lambda x: new_desc(x))
```

```python
data['Deslocamento'].unique()
```

### Corrigindo coluna 'mod_força'

```python
# mod forçaestá sem valores negativos
data[data['mod_força'].isin([1])]
```

```python
# corrigindo mod força

for i in range(len(data)):
    if data.loc[i,'Força'] < 10:
        x = -1*abs(data.loc[i, 'mod_força'])
        data.loc[i, 'mod_força'] = x
    else:
        data.loc[i, 'mod_força'] = data.loc[i, 'mod_força']
```
