```python
# unindo as informações das 2 páginas em uma única lista

zipado = zipado1 + zipado2
len(zipado)
```

```python
# unindo páginas a serem pesquisadas

total_monsters = paginas1 + paginas2
total_monsters = list((map(int, total_monsters)))
print('Número total de monstros:', len(total_monsters))
```

```python
# removendo páginas repetidas

paginas = list(set(total_monsters))
```

```python
# extraindo informações de cada uma das páginas solicitadas

for i in tqdm(range(len(paginas))):
    # realizando leitura da página
    pageObj = pdfReader.getPage(paginas[i])
    page = pageObj.extractText()
    
    # arrumando texto
    textonovo = page.replace('\n', '')
    
    # verificando se o item está na página e adicionando seu nome a uma lista
    names = []
    for k in range(len(zipado)):
        if int(zipado[k][1]) == paginas[i]:
            names.append(zipado[k][0])
    
    # printa a lista de nomes a serem extraídos da página
    print(names)
    
    # contagem de itens na lista de nomes a serem extraídos
    counter = len(names)
    counter
    
    # pra cada item na lista de nomes a serem extraídos da página, extrai as 
    # informações e printa na tela
    # ao final adiciona as informações no dataframe principal
    for j in range(counter):
        monster = names[j]
        print('Nome:', monster)
        
        try:
            monster_txt = re.findall(f'{monster} (\w+ \w+\s*[)(\w]*, .+)', textonovo)[0]
        except:
            monster_txt = textonovo

        tamanho = re.findall(r'[A-Z][\w-]+ ((?:Pequen[oa])|(?:Médi[oa])|(?:Grande)|(?:Enorme)|(?:Imens[oa])|(?:Miúd[oa]))[\s\w)(-]*,', monster_txt)[0]
        print('Tamanho:', tamanho)            

        tipo = re.findall(f'([A-Z][\w-]+) {tamanho}[\s\w)(-]*,', monster_txt)[0]
        print('Tipo:', tipo)

        tendencia = re.findall(f'{tamanho}[\s\w,)(-]*, ([\w\s]+) (?:[A-Z(]|não)', monster_txt)[0]
        print('Tendência:', tendencia)

        CA = re.findall(r'Classe de Armadura (\d+)', monster_txt)[0]
        print('CA:', CA)

        PV_avg = re.findall(r'Pontos de Vida (\d+)', monster_txt)[0]
        print('Pontos médio de vida:', PV_avg)

        PV_dice = re.findall(r'Pontos de Vida \d+ [(]([\w\d\s+]+)[)]', monster_txt)[0]
        print('Pontos de vida - dados:', PV_dice)

        desloc = re.findall(r'Deslocamento ([\d,]+ m)', monster_txt)[0]
        print('Deslocamento', desloc[0])
        
        try:
            FOR = re.findall(r'FOR (\d+)', monster_txt)[0]
            print('Força:', FOR)
            FOR_mod = re.findall(r'FOR \d+ [(]([+-]*\d+)', monster_txt)[0]
            print('Modificador de força:', FOR_mod)
        except:
            FOR = np.nan
            FOR_mod = np.nan
        
        try:
            DES = re.findall(r'DES (\d+)', monster_txt)[0]
            print('Destreza:', DES)
            DES_mod = re.findall(r'DES \d+ [(]([+-]*\d+)', monster_txt)[0]
            print('Modificador de Destreza:', DES_mod)
        except:
            DES = np.nan
            DES_mod = np.nan  
        
        try:
            CON = re.findall(r'CO[NM] (\d+) [(]', monster_txt)[0]
            print('Constituição:', CON)
            CON_mod = re.findall(r'CO[NM] \d+ [(]([+-]*\d+)', monster_txt)[0]
            print('Modificador de Constituição:', CON_mod)
        except:
            CON = np.nan
            CON_mod = np.nan
            
        try:
            INT = re.findall(r'INT (\d+)', monster_txt)[0]
            print('Inteligência:', INT)
            INT_mod = re.findall(r'INT \d+ [(]([+-]*\d+)', monster_txt)[0]
            print('Modificador de Inteligência:', INT_mod)
        except:
            INT = np.nan
            INT_mod = np.nan
            
        try:
            SAB = re.findall(r'SAB (\d+)', monster_txt)[0]
            print('Sabedoria:', SAB)
        except:
            SAB = np.nan
        
        try:
            SAB_mod = re.findall(r'SAB \d+ [(]([+-]*\d+)', monster_txt)[0]
            print('Modificador de Sabedoria:', SAB_mod)
        except:
            SAB_mod = np.nan
            
        try:
            CAR = re.findall(r'CAR (\d+)', monster_txt)[0]
            print('Carisma:', CAR)
            CAR_mod = re.findall(r'CAR \d+ [(]([+-]*\d+)', monster_txt)[0]
            print('Modificador de Carisma:', CAR_mod)
        except:
            CAR = np.nan
            CAR_mod = np.nan
        
        try:
            infos = re.findall(r'CAR \d+ .[+-]*\d+. (.+) Nível', monster_txt)[0]
            print('Informações:', infos)
        except:
            infos = np.nan
            
        try:
            nivel = re.findall(r'Desafio (\d+[/\d]*)', monster_txt)[0]
            print('Nível de desafio:', nivel)
        except:
            nivel = np.nan
            
        try:
            xp = re.findall(r'Desafio \d+[/\d]* .([\d.]+)', monster_txt)[0]
            print('XP:', xp)
        except:
            xp = np.nan
        
        try:
            acoes = re.findall(r'AÇÕES([\d\D]{0,600} dano [\w\s]{0,30}[.]) ', monster_txt)[0]
            print('Ações:', acoes)
        except:
            acoes = np.nan
        
        def dano_extra(monster_txt):
            try:
                x = int(re.findall(r'(?:Acerto: .+ mais (\d+)) ', monster_txt)[0])
                return x
            except:
                return 0
        try:
            dano = max(list(map(int, re.findall(r'(?:Acerto: (\d+))', monster_txt)[:2]))) + dano_extra(monster_txt)
            print('Dano médio:', (dano))
        except:
            try:
                dano = re.findall('(\d+) [(]', acoes)[0]
            except:
                dano = np.nan

        df_monster = df_monster.append({'Nome':monster, 'Tipo':tipo, 'Tamanho':tamanho, 'Tendência':tendencia, 'PV_médio':PV_avg, 'PV_dados':PV_dice, 'CA':CA, 'Deslocamento':desloc,
                              'Força':FOR, 'mod_força':FOR_mod, 'Destreza':DES, 'mod_destreza':DES_mod, 'Constituição':CON, 'mod_constituição':CON_mod,
                              'Inteligência':INT, 'mod_inteligência':INT_mod, 'Sabedoria':SAB, 'mod_sabedoria':SAB_mod, 'Carisma':CAR,
                              'mod_carisma':CAR_mod, 'Infos':infos, 'Nível_desafio':nivel, 'XP':xp, 'Ações':acoes, 'Dano_médio':(dano)}, ignore_index = True)

        print('\n\n')
```
