```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from lightgbm import LGBMClassifier
import lightgbm

from yellowbrick.model_selection import LearningCurve
from yellowbrick.classifier import confusion_matrix as cm
from yellowbrick.classifier import ROCAUC

from sklearn.metrics import roc_auc_score
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score
```

```python
data = pd.read_csv('../simulation/battle_simulation.csv')
data.head()
```

# LogisticRegresion

```python
X = data.drop(columns = 'resultado_battle')
y = data['resultado_battle']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Standardzation dos dados para utilziar na regressão logística

std = StandardScaler()
X_std = std.fit_transform(X)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# regressão logística
logistic = LogisticRegression(max_iter = 1000)

#sepparando em teste e treino
X_train, X_test, y_train, y_test = train_test_split(X_std, y)

logistic.fit(X_train, y_train)
y_pred_test = logistic.predict(X_test)
y_pred_train = logistic.predict(X_train)

#resultados treino
print('Train')
print('accuracy:', accuracy_score(y_train, y_pred_train))
print('precision:', precision_score(y_train, y_pred_train))
print('recall:', recall_score(y_train, y_pred_train))
print(classification_report(y_train, y_pred_train))
print(roc_auc_score(y_train, logistic.predict_proba(X_train)[:,1]))

#resultados teste
print('------------------------\n','Test')
print('accuracy:', accuracy_score(y_test, y_pred_test))
print('precision:', precision_score(y_test, y_pred_test))
print('recall:', recall_score(y_test, y_pred_test))
print(classification_report(y_train, y_pred_train))
print(classification_report(y_test, y_pred_test))
print(roc_auc_score(y_test, logistic.predict_proba(X_test)[:,1]))
```

```python
from sklearn.model_selection import cross_val_score, KFold
```

```python
# cross validation
cross_val_score(logistic, X, y, cv = KFold(10), scoring = 'recall')
```

```python
# visualização da matriz de confusão
visualizer = cm(logistic, X_test, y_test, test_size=0.999)
```

```python
# analisando distribuição das probabilidades
plt.hist(logistic.predict_proba(X_test)[:,1]);
```

```python
# plotando a curva de aprendizagem

visualizer = LearningCurve(logistic)
visualizer.fit(X_std, y)
visualizer.show()
```

```python
# plotando a curva roc

visualizer = ROCAUC(logistic)

visualizer.fit(X_train, y_train)        # Fit the training data to the visualizer
visualizer.score(X_test, y_test)        # Evaluate the model on the test data
visualizer.show()                       # Finalize and show the figure
```

# LightGBM

```python
# separando em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y)

# modelo lightgbm
light = LGBMClassifier(max_depth=3)#max_depth=5,num_leaves=20, n_estimators=150,)# colsample_bytree=0.7)
light.fit(X_train, y_train)
y_pred_test = light.predict(X_test)
y_pred_train = light.predict(X_train)


#resultados no treino
print('Train')
print('accuracy:', accuracy_score(y_train, y_pred_train))
print('precision:', precision_score(y_train, y_pred_train))
print('recall:', recall_score(y_train, y_pred_train))
print(classification_report(y_train, y_pred_train))
print(roc_auc_score(y_train, light.predict_proba(X_train)[:,1]))

#resultados no teste
print('------------------------\n','Test')
print('accuracy:', accuracy_score(y_test, y_pred_test))
print('precision:', precision_score(y_test, y_pred_test))
print('recall:', recall_score(y_test, y_pred_test))
print(classification_report(y_train, y_pred_train))
print(classification_report(y_test, y_pred_test))
print(roc_auc_score(y_test, light.predict_proba(X_test)[:,1]))
```

```python
# cross validation do modelo
cross_val_score(light, X, y, cv = KFold(10), scoring='recall')
```

```python
from sklearn.metrics import plot_confusion_matrix
```

```python
# plotando a matriz de confusão
plot_confusion_matrix(light, X_test, y_test, values_format='.0f');
```

```python
# plotando a importancia das features levando em consideração o ganho

lightgbm.plot_importance(light, importance_type='gain', max_num_features = 15)
```

```python
# analisando distribuição das probabilidades

plt.hist(light.predict_proba(X_test)[:,1]);
```

```python
# plotando a curva de aprendizagem

visualizer = LearningCurve(light)
visualizer.fit(X, y)
visualizer.show()
```

```python
# plotando a curva roc

visualizer = ROCAUC(light)

visualizer.fit(X_train, y_train)        # Fit the training data to the visualizer
visualizer.score(X_test, y_test)        # Evaluate the model on the test data
visualizer.show()                       # Finalize and show the figure
```

```python
# analisando algumas features para possível diminuição do número de features em possível deploy

from yellowbrick.features import RFECV

visualizer = RFECV(light, n_jobs = -1)
visualizer.fit(X_train[['n_players', 'max_pv_player', 'max_ca_player', 'max_dano_player',
                        'max_con_player', 'max_for_player',
                        'n_creature', 'max_pv_creature', 'max_ca_creature', 'max_dano_creature',
                        'max_con_creature', 'max_for_creature']], y_train)
visualizer.show()
```
